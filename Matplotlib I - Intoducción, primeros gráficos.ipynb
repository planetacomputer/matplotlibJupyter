```python
# Imaginemos que es un gr치fico en blanco

fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

axes.plot(x,y)
axes.set_xlabel('Variable X')
axes.set_ylabel('Variable Y')
axes.set_title('T칤tulo')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Gr치fico grande')
axes1.set_xlabel('Variable del eje X')
axes1.set_ylabel('Variable   del eje Y')

axes2.plot(y,x)
axes2.set_title('Gr치fico peque침o')
axes2.set_xlabel('Variable del eje X')
axes2.set_ylabel('Variable del eje Y')
```
