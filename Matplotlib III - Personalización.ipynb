```python
# Liberías

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Variables

x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color = '#FF8C00', # Códigos de colores RGB HEX y nombres de colores (purple, green)
                linewidth = 2, # anchura de la línea
                alpha = 0.5, # transparencia 
                linestyle = '--', # estilo de la línea, continua, puntos... (:, steps, -)
                marker = 'o', # marca en cada punto del array
                markersize = 20, # tamaño de la marca de cada punto del array
                markerfacecolor = 'yellow', # color de la marca de cada punto del array
                markeredgewidth = 1, # tamaño del borde de la marca de cada punto del array
                markeredgecolor = 'green'
                
       )  
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # color y transparencia  de la línea
ax.plot(x, x+2, color="#8B008B")        # Código RGB hex 
ax.plot(x, x+3, color="#FF8C00")        # código RGB hex  
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Por defecto")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("Ajustado")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Personalización de ejes");
```
