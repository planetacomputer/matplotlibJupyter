```python
## Matplotlib II
## Pixeles, leyendas y gr√°ficos
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Gr√°fico grande')
axes1.set_xlabel('Variable del eje X')
axes1.set_ylabel('Variable   del eje Y')

axes2.plot(y,x)
axes2.set_title('Gr√°fico peque√±o')
axes2.set_xlabel('Variable del eje X')
axes2.set_ylabel('Variable del eje Y')
```

```python
fig, axes = plt.subplots(nrows= 2, ncols = 1, figsize=(5,3))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig
```

```python
# Guardando el gr√°fico como una imagen
fig.savefig('my_first_picture_dpi.png', dpi = 200) # dpi hace referencia a los pixeles
```

```python
# Trabajando con la leyenda
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x **2, label = 'X al cuadrado')
ax.plot(x,x **3, label = 'Y al cuadrado')

ax.legend(loc = 'best')
```
