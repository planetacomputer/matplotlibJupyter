## Ejercicio 1

**Sigue los siguientes pasos:**
* **Crea una figura usando el comando plt.figure()**
* **Usa add_axes para añadir un eje a la figura en [0,0,1,1]. Llama a este nuevo eje ax.**
* **Visualiza (x,y) en los ejes y pon las etiquetas y los títulos al gráfico:**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

plt.plot(x,y,'b')
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.set_title('titulo')
```

## Ejercicio 3

**Crea una figura que contenga un gráfico en [0,0,1,1] y otro en la misma figura en los ejes [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

**Ahora usa x, y, z para visualizar los valores en el gráfico. El gráfico con los ejes [0.2,0.5,.4,.4] tiene aparentar que estamos haciendo zoom sobre el gráfico [0,0,1,1]. ¡Lo utilizaremos como una lupa!**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z)
ax1.set_ylabel('Z')
ax1.set_xlabel('X')

ax2.plot(x,y)
ax2.set_ylabel('Y')
ax2.set_xlabel('X')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

## Ejercicio 4

**Usa plt.subplots(nrows=1, ncols=2) para crear dos gráficos.**

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

**Ahora dibuja (x,y) y (x,z) en los ejes y personaliza los gráficos**

```python
axes[0].plot(x, y, color='b',lw=3, ls='--')
axes[1].plot(x, z, color='r', lw=3)
fig
```

**Utiliza figsize() como argumento en plt.subplots()**

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(12,5))

axes[0].plot(x, y, color='b',lw=3, ls='--')
axes[0].set_ylabel('y')
axes[0].set_xlabel('x')

axes[1].plot(x, z, color='r', lw=3)
axes[0].set_ylabel('z')
axes[0].set_xlabel('x')
```
